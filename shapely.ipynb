### Drawing and Creating Shapes using Shapely

Shapely is a handy python package that easily creates geometric figures and combines them into new ones.  Shapely has a handy library of functions.  To learn more, check out Shapely's manual:

http://toblerity.org/shapely/manual.html

Let's get started!

Import the packages you will need to draw things

```python
import draw_something

from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
import pysal
import pylab as pl
import shapely
import shapely.geometry
from shapely.geometry import shape
from shapely.geometry import Point, LineString, Polygon, MultiPolygon 
```

Create a point and draw a radius around it to create a circle

```python
pt = Point(0, 0)
circle = pt.buffer(.5)
draw_something.plot([circle, pt])

```

Create and draw a second circle on top

```python
another_pt = Point(0, .25)
another_circle = another_pt.buffer(.5)
draw_something.plot([circle, another_circle], color_dict=None)
```

You can use union to combine two shapes into one figure

```python
merged_shape = another_circle.union(circle)
draw_something.plot(merged_shape, color_dict=None)
```

You can also use 'intersection' to create a shape that is the area overlapped by both shapes

```python
intersect_shape = another_circle.intersection(circle)
draw_something.plot(intersect_shape, color_dict=None)
```

Shapely is very easy to use!
